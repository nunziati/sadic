```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("proteins_dataset_PDB_with_protein_size.csv")
df2 = pd.read_csv("protein_sample_40000.txt")
x = df[df["PDB ID"].isin(df2["pdb_id"])]["N"]

f = plt.figure()
ax = f.add_subplot(111)

ax.hist(x, bins=30, edgecolor="black")
ax.set_xlabel("Number of atoms")
ax.set_ylabel("Frequency")
ax.set_xticks([0, 10000, 20000, 30000, 40000])

ax.xaxis.label.set_size(16)
ax.yaxis.label.set_size(16)
ax.tick_params(axis='both', labelsize=13)
f.tight_layout()

f.savefig("N_distribution_sample.pdf")

plt.show()
```
